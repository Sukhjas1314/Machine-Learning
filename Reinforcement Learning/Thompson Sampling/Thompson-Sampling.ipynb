# Thompson-Sampling Model

### Importing libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Importing the dataset

```python
data = pd.read_csv('Ads_CTR_Optimisation copy.csv')
```

### Implementing the Thompson Sampling model

```python
import random
N = 500
d = 10
ads_selected = []
no_of_rewards_1 = [0] * d
no_of_rewards_0 = [0] * d
total_reward = 0
for n in range(0,N):
    ad = 0
    max_random = 0
    for i in range(0,d):
        random_beta = random.betavariate(no_of_rewards_1[i] + 1,no_of_rewards_0[i] + 1)
        if(random_beta > max_random):
            max_random = random_beta
            ad = i
    ads_selected.append(ad)
    reward = data.values[n,ad]
    if(reward == 1):
        no_of_rewards_1[ad] = no_of_rewards_1[ad] + 1
    else: 
        no_of_rewards_0[ad] = no_of_rewards_0[ad] + 1 
    total_reward = total_reward + reward
```

### Visualising the results

```python
plt.hist(ads_selected)
plt.title('Histogram of Ads Selections')
plt.xlabel('Ads')
plt.ylabel('Number of times each ad was selected')
plt.show()
```
