# Decision Tree

### Importing libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Importing the dataset

```python
df = pd.read_csv("Position_Salaries.csv")
X = df.iloc[:,1:-1].values
y = df.iloc[:,-1].values
```

### Training Decision Tree Regression model on whole dataset

```python
from sklearn.tree import DecisionTreeRegressor
dr = DecisionTreeRegressor(random_state = 0)
dr.fit(X,y)
```

### Predicting the result

```python
print(dr.predict([[6.5]]))
```

### Visualising Decision Tree Regression results (higher resolution)

```python
X_grid = np.arange(min(X),max(X),0.1)
X_grid = X_grid.reshape((len(X_grid),1))
plt.scatter(X,y,color = "red")
plt.plot(X_grid,dr.predict(X_grid),color = "blue")
plt.title("Truth or Bluff (Decision Tree regression)")
plt.xlabel("Position level")
plt.ylabel("Salary")
plt.show()
```
